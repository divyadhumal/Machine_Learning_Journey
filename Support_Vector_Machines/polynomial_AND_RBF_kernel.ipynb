```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
#create a Toy sataswt using make_moons

from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.2, random_state=42)
```

```python
#split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state = 42)
```

```python
from sklearn.svm import SVC

#create svm model with polynomial kernel
poly_model = SVC(kernel='poly',degree=3)
poly_model.fit(X_train, y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#create a meshgrid to plot
x_min, x_max = X[:,0].min() - 1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx,yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                    np.linspace(y_min, y_max,500))

#predict on the grid
Z= poly_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z =Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contour(xx, yy, Z, alpha=0.3, cmap = 'coolwarm')
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', s=30)
plt.title("SVM with Polynomial Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
#create SVM model with RBF kernel
rbf_model = SVC(kernel='rbf')
rbf_model.fit(X_train, y_train)
```

```python
x_min , x_max= X[:,0].min()-1, X[:,0].max()+1
y_min, y_max= X[:,1].min()-1, X[:,1].max()+1

XX,YY = np.meshgrid(np.linspace(x_min, x_max, 500),
                    np.linspace(y_min, y_max,500))

#predict on the grid
Z = rbf_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z= Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contour(XX, YY, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', s=30)
plt.title("SVM with RBF Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

#### compare Polynomial vs RBF Kernel

```python
#Accuracy for Polynomial
poly_pred = poly_model.predict(X_test)
poly_acc = accuracy_score(y_test, poly_pred)

#Accuracy for RBF
rbf_pred = rbf_model.predict(X_test)
rbf_acc = accuracy_score(y_test, rbf_pred)

print("Polynomial Kernel Accuracy:", poly_acc)
print("RBF Kernel Accuracy:", rbf_acc)
```
