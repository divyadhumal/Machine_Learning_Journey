```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\DIVYA\Downloads\WA_Fn-UseC_-Telco-Customer-Churn.csv")
print(df.head())
```

```python
df['Churn'].value_counts()
```

```python
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
df.groupby('Churn')['tenure'].mean().plot(kind='bar')
plt.show()
```

#### New customers leave more, long-term customers stay

```python
#are more customers staying or leaving
df['Churn'].value_counts().plot(kind='bar')
plt.title("Churn Count")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df['tenure'].plot(kind='hist', bins=20)
plt.title("Tenure Distribution")
plt.xlabel("Tenure")
plt.show()
```
