```python
#Boundary Extraction
import matplotlib.pyplot as plt
import numpy as np

b0 = -2
b1 = 0.03
b2 = 0.5

#range of ages
age = np.linspace(20,60,100)

#Calculate the income that would put the logit(p)=0 (decision boundary line)

income_boundary = (-b0 -b1*age)/b2

np.random.seed(0)
ages = np.random.uniform(20, 60, 50)
incomes = np.random.uniform(1,5,50)

logits = b0 +b1*ages + b2*incomes
labels = (logits >= 0).astype(int)

plt.figure(figsize= (8,6))
plt.scatter(ages[labels ==0], incomes[labels==0], color='blue', label ="Class 0 ('No')")
plt.scatter(ages[labels ==1], incomes[labels ==1], color ='green', label = "Class 1 ('Yes')")
plt.plot(age, income_boundary, color ='Red', label="Decision Boundary(p = 0.5)")
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Decision Boundary for Logistic Regression (2D example)")
plt.legends()
plt.grid(True)
plt.show()
```
